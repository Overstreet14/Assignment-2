```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Load the data
kevin = pd.read_csv('/content/dft-road-casualty-statistics-casualty-provisional-mid-year-unvalidated-2022 (1).csv')
# Extract the independent variable (X) and dependent variable (y)
X = kevin[['casualty_severity']]
y = kevin['age_of_casualty']
# Create and fit a linear regression model
regressor = LinearRegression()
regressor.fit(X, y)
# Plot the data points
kevin.plot(style='o',x='casualty_severity',y='age_of_casualty', label='Data Points')
# Overlay the regression line
plt.plot(X, regressor.predict(X), color='red', linewidth=2, label='Regression Line')
plt.xlabel('casualty_severity')
plt.ylabel('age_of_casualty')
plt.legend()
plt.show()
```
